```python
### Helper Function for adding the hits across haplotypes
### What this means is that we transform a DataFrame of peptides vs alleles to a DataFrame of peptides vs haplotypes
def add_hits_across_haplotypes(df_binarized_final, df_hap_freq):
    
    df_overlap = pd.DataFrame(index = df_binarized_final.index, columns = df_hap_freq.columns)
    
    unique_columns = df_binarized_final.columns.unique()
    count = 0
    print(unique_columns)
    for pept in df_overlap.index:
        count += 1
        if count%50 == 0:
            print('Round = ', count)
        for col in df_overlap.columns:
            temp_sum = 0
            for hap_type in col:
                if hap_type in unique_columns:
                    temp_sum += int(df_binarized_final.loc[pept, hap_type])
    
            df_overlap.loc[pept, col] = temp_sum
    
    return df_overlap
```

```python
### Load the file containing the table of peptide vs allele predicted binding affinity
mhc1_mhcflurry_pred_affinity = pd.read_pickle('mhc1_haplotype_mhcflurry_pred_affinity_pivot.pkl.gz')
```

```python
mhc1_mhcflurry_pred_affinity
```

```python
### Binarize the data to create hits
mhc1_mhcflurry_pred_affinity_binarized = mhc1_mhcflurry_pred_affinity.applymap(lambda x: 1 if x>0.638 else 0)
```

```python
mhc1_mhcflurry_pred_affinity_binarized
```

```python
### Sum across peptides to see which are useless, i.e. have no hits whatsoever
mhc1_mhcflurry_pred_affinity_binarized.sum(axis=1)
```

```python
### Peptides in MHC Class I that have zero hits
zero_hit_peptides_mhc1 = mhc1_mhcflurry_pred_affinity_binarized.sum(axis=1)[mhc1_mhcflurry_pred_affinity_binarized.sum(axis=1) == 0].index
```

```python
zero_hit_peptides_mhc1
```

```python
### Get rid of the zero hit peptides
mhc1_mhcflurry_pred_affinity_binarized.drop(zero_hit_peptides_mhc1, axis='index', inplace=True)
```

```python
mhc1_mhcflurry_pred_affinity_binarized
```

```python
mhc1_mhcflurry_pred_affinity_binarized.sum(axis=1)
```

```python
### Drop the loci from the columns and get the data down to a Single Index of alleles for the columns
mhc1_mhcflurry_pred_affinity_binarized = mhc1_mhcflurry_pred_affinity_binarized.droplevel('loci', axis=1)
```

```python
mhc1_mhcflurry_pred_affinity_binarized
```

```python
### Run the add_hits_across_haplotypes function on the desired data.
overlap_haplotypes_mhc1 = add_hits_across_haplotypes(mhc1_mhcflurry_pred_affinity_binarized, hap_freq_mhc1)
```

```python
overlap_haplotypes_mhc1
```

```python
### Make sure that all the peptides have some hits - A sanity check.
overlap_haplotypes_mhc1.sum(axis=1)
```
