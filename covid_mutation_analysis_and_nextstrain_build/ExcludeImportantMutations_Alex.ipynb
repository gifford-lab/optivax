### South African Mutation (501Y.V2)

```python
### South African Mutation, add a column of binary values showing whether there is a mutation there
my_df['South_African_mutation'] = pd.Series(np.zeros(len(my_df)))
```

```python
my_df
```

```python
south_african_indeces = my_df[np.logical_or(my_df.protein=='S1', my_df.protein=='S2')].index
```

```python
south_african_indeces
```

```python
def south_africa(mutation_point):
    '''
    Inputs:
        - mutation point: the number of the nucleotide in the S protein that is mutated
    Outputs:
        nothing
        But, it changes the my_df dataframe
    '''
    for my_index in south_african_indeces:
        temp_start = my_df.loc[my_index, 'start_pos']
        temp_len = my_df.loc[my_index, 'epi_len']
        if temp_start <= mutation_point and temp_start+temp_len > mutation_point:
            my_df.loc[my_index, 'South_African_mutation'] = 1
#             print(my_df.loc[my_index,:])
        
```

```python
def uk_mutation(mutation_point, my_protein):
    '''
    Inputs:
        - mutation point: the number of the nucleotide in the S protein that is mutated
        - protein: the protein, ex. S (S1 and S2), ORF1a (there is no need to say ORF1ab as 
        there are no mutations past the cutoff between ORF1a and ORF1b) etc.
    Outputs:
        nothing
        But, it changes the my_df dataframe
    '''
    if my_protein == 'S':
        indeces = my_df[np.logical_or(my_df.protein=='S1', my_df.protein=='S2')].index
    else:
        indeces = my_df[my_df.protein==my_protein].index
        
    for my_index in indeces:
        temp_start = my_df.loc[my_index, 'start_pos']
        temp_len = my_df.loc[my_index, 'epi_len']
        if temp_start <= mutation_point and temp_start+temp_len > mutation_point:
            my_df.loc[my_index, 'UK_mutation'] = 1
#             print(my_index)
#             print(my_df.loc[my_index,:])
```

```python
### From this paper: https://virological.org/t/preliminary-genomic-characterisation-of-an-emergent-sars-cov-2-lineage-in-the-uk-defined-by-a-novel-set-of-spike-mutations/563
### 0-indexed!!!, also instead of ORF1ab I use ORF1a, because all the mutations are in the beginning of 
### the ORF1ab, so in ORF1a

### ORF1ab -> ORF1a
### T1001I
uk_mutation(1000, 'ORF1a')
### A1708D
uk_mutation(1707, 'ORF1a')
### I2230T
uk_mutation(2229, 'ORF1a')
### SGF 3675-3677 deletion
uk_mutation(3674, 'ORF1a')
uk_mutation(3675, 'ORF1a')
uk_mutation(3676, 'ORF1a')

### S (Spike protein)
### HV 69-70 deletion
uk_mutation(68, 'S')
uk_mutation(69, 'S')
### Y144 deletion
uk_mutation(143, 'S')
### N501Y
uk_mutation(500, 'S')
### A570D
uk_mutation(569, 'S')
### P681H
uk_mutation(680, 'S')
### T716I
uk_mutation(715, 'S')
### S982A
uk_mutation(981, 'S')
### D1118H
uk_mutation(1117, 'S')

### ORF8
### Q27stop -> Basically get rid of all of ORF8 !!!!!

# ### R52I
# uk_mutation(51, 'ORF8')
# ### Y73C
# uk_mutation(72, 'ORF8')

### N
### D3L
uk_mutation(2, 'N')
### S235F
uk_mutation(234, 'N')
```

### Indian Mutation (B.1.617)

```python
### Use paper: https://www.biorxiv.org/content/10.1101/2021.04.23.441101v1.full
```

```python
my_df = pd.read_csv('data/processed/AllEpitopeFeatures_with_South_African_and_UK_mutation_larger.csv')
```

```python
my_df
```

```python
my_df['Indian_mutation'] = pd.Series(np.zeros(len(my_df)))
```

```python
def indian_mutation(mutation_point, my_protein):
    '''
    Inputs:
        - mutation point: the number of the nucleotide in the S protein that is mutated
        - protein: the protein, ex. S (S1 and S2), ORF1a, ORF1b etc.
    Outputs:
        nothing
        But, it changes the my_df dataframe
    '''
#     if my_protein == 'S':
#         indeces = my_df[np.logical_or(my_df.protein=='S1', my_df.protein=='S2')].index
#     elif my_protein == '':
#         indeces = my_df[my_df.protein==my_protein].index

    indeces = my_df[my_df.protein==my_protein].index
    
    anything = False
        
    for my_index in indeces:
        temp_start = my_df.loc[my_index, 'start_pos']
        temp_len = my_df.loc[my_index, 'epi_len']
        if temp_start <= mutation_point and temp_start+temp_len > mutation_point:
            my_df.loc[my_index, 'Indian_mutation'] = 1
            anything = True
            
    if not anything:
        print('Problem with ', mutation_point, ' in protein ', my_protein)
#             print(my_index)
#             print(my_df.loc[my_index,:])
```

```python
### Mutation points for the Indian Mutation (B.1.617)
### From the paper: https://www.biorxiv.org/content/10.1101/2021.04.23.441101v1.full
### For all the proteins the mutation point is -1 from what is written, because of zero indexing
### Exception: ORF1b, where the mutation point is -3 - 4399, because the numbering continues over from ORF1a
### Note that we need not worry in the same way about S2, as the numbering simply continues on there.

### ORF1a
### T 1567 I 
indian_mutation(1567 - 1, 'ORF1a')
### T 3646 A
indian_mutation(3646 - 1, 'ORF1a')

### ORF1b 
### P 4715 L
indian_mutation(4715 - 3 - 4399, 'ORF1b')
### G 5530 I
indian_mutation(5530 - 3 - 4399, 'ORF1b')
### M 5753 I
indian_mutation(5753 - 3 - 4399, 'ORF1b')
### K 6711 R
indian_mutation(6711 - 3 - 4399, 'ORF1b')
### S 6713 A
indian_mutation(6713 - 3 - 4399, 'ORF1b')

### S protein
### S1 domain
### T 95 I
indian_mutation(95 - 1, 'S1')
### G 142 D
indian_mutation(142 - 1, 'S1')
### E 154 K
indian_mutation(154 - 1, 'S1')
### L 452 R
indian_mutation(452 - 1, 'S1')
### E 484 Q
indian_mutation(484 - 1, 'S1')
### D 614 G
indian_mutation(614 - 1, 'S1')
### P 681 R
indian_mutation(681 - 1, 'S1')
### S2 domain
### Q 1071 H
indian_mutation(1071 - 1, 'S2')

### ORF3a
### S 26 L
indian_mutation(26 - 1, 'ORF3a')

### M
### I 82 S
indian_mutation(82 - 1, 'M')

### ORF6
### D 61 (the last amino acid)
indian_mutation(61 - 1, 'ORF6')
### 62 (anything after that)

### ORF7a
### N 43 Y
indian_mutation(43 - 1, 'ORF7a')
### V 82 A
indian_mutation(82 - 1, 'ORF7a')

### N
### R 203 M
indian_mutation(203 - 1, 'N')
### D 377 Y
indian_mutation(377 - 1, 'N')
```

### Set the mutation rate equal to 1 if the peptide is hit by (South_African, UK, Indian, Brazilian) mutation

```python
my_df = pd.read_csv('data/processed/AllEpitopeFeatures_with_South_African_and_UK_and_Indian_and_Brazilian_mutation_larger.csv')
```

```python
my_df
```

```python
for ind in my_df.index:
    if ind%10000==0:
        print(ind)
    if my_df.loc[ind,'UK_mutation'] == 1 or my_df.loc[ind,'South_African_mutation'] == 1 or my_df.loc[ind,'Indian_mutation'] == 1 or my_df.loc[ind,'Brazilian_mutation'] == 1:
        my_df.loc[ind,'perc_mutated'] = 1.0
```

```python
new_df = my_df.drop(labels = ['South_African_mutation','UK_mutation', 'Indian_mutation', 'Brazilian_mutation'], axis='columns')
```
